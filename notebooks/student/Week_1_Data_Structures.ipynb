````python
# Sets
a = {1,2,3,4,5}
b = {4,5,6,7,8}
print("a: ", a)
print("b: ", b)

print("A Union B: all unique values combined")
print(a | b) # A Union B
print(a.union(b)) # A Union B

print("A Intersection B: values they have in common")
print(a & b) # A INtersection B
print(a.intersection(b)) # A INtersection B

print("Difference: values in common from B removed from A")
print(a.difference(b)) # A - B
print("Difference: values in common from A removed from B")
print(b.difference(a)) # B - A

print("Symmetric Difference: Values in common removed then combining both remaining sets")
print(a.symmetric_difference(b)) # (A U B) - (A INtersection B)
````

Given:

a = {1, 2, 3, 4}

b = {3, 4, 5, 6}

Find:

- common elements
- all unique elements together
- elements in a but not in b

## NOTE: SETS are unordered, so Python may display them in a different order. For small integers it often looks sorted, but you should not rely on that.

````python
# SETS are unordered, so Python may display them in a different order. For small integers it often looks sorted, but you should not rely on that.

a = {1, 2, 3, 4}

b = {3, 4, 5, 6}

print("common elements")
common_elements = a & b
print(common_elements, "\n")
print("all unique elements together")
unique_elements = a | b
print(unique_elements, "\n")
print("elements in 'a' but not in 'b'")
a_not_in_b = a.difference(b)
print(a_not_in_b)
````
