```python
len(a[:5]) # To check the length of the string
```

## String Methods

```python
a = "Hello World       "
#lower()
#upper()
#rstrip() strip white space on right of the string
#lstrip() strip white space on left of the string

print(len(a))
print(a.lower())
print(a.upper())
print(a.rstrip())
print(len(a))
```

```python
e = "bmaunmdbraai"
print(e[0:len(e):2])
print(e[1:len(e):2])
# or you can omit len(e) because it will go to the end
print(e[0::2])
print(e[1::2])
```
