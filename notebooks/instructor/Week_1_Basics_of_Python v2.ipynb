In a certain encrypted message which has information about the location (area,city) the characters are jumbled such that the first character of the first word is followed by the first character of the second word, then it is followed by second character of the first word and so on....

In other words, lets say the location is
bandra,mumbai

The encrypted message says 'bmaunmdbraai'

- Sample input:
bmaunmdbraai
- Sample output:
bandra,mumbai

```python
enc = 'bmaunmdbraai'
area = enc[::2]
city = enc[1::2]

print(area)
print(city)
```
