## **Type Casting**

- int to float
- float to int
- float to string
- string to float/int

```python
a = "5"
b = "6"

print(int(a) + int(b))
```

```python
a = "5"
b = "6.5"

print(float(a) + float(b))
```

```python
#extend
#append

list1 = [1,2,3,4]
list2 = [5,6,7,8]

print(list1+list2)
list1.append(list2)
print(list1)
```

```python
#extend
#append

list1 = [1,2,3,4]
list2 = [5,6,7,8]

print(list1+list2)
list1.extend(list2)
print(list1)
```

## **Sets**

 - defined as {}
 - Doesn't store duplicate elements


```python
a = [1,2,3,4,5,5,2,4,2,5,2,1,4,2,2]

len(a)
```

```python
b = set(a)
print(b)

print(b[2])
```

```python
## Ven Diagram

a = {1,2,3,4,5,6}
b = {2,4,6,8,10}
```

```python
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```
